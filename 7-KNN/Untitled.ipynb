# KNN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data.csv")
data
```

```python
data.info()
```

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace=True)
data.tail()
```

```python
M=data[data.diagnosis=="M"] #kötü huylu tümör
B=data[data.diagnosis=="B"]  # iyi huylu tümör
```

```python
#scatter 
plt.scatter(M.radius_mean,M.texture_mean,color="red",label="Kotu")
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="iyi")
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.show()
```

```python
# M ve B değerleri ya kategorik yada int olmalı

#int yapalım

data.diagnosis=[1 if each=="M" else 0 for each in data.diagnosis]
```

```python
# x=feature larım
# y=class  larım

y=data.diagnosis.values  #diagnosis class
x=data.drop(["diagnosis"],axis=1)  #feature lar 
```

```python
x
```

```python
y
```

```python
#normalization

x=(x-np.min(x))/(np.max(x)-np.min(x))
```

```python
x
```

```python
#train ve test 

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
## knn model

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)  # k=3
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
```

```python
prediction
```

```python
print(" {} nn score: {} ".format(3,knn.score(x_test,y_test)))
```

En uygun k değeri nasıl bulunur Buradaki k değeri bir hyperparameter olduğundan en uygun olanını bulalım

```python
#find k value
score_list=[]
for each in range(1,15):
    knn2=KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    knn2.predict(x_test)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

```python
#En uygun değer 6-10

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)  # k=3
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
print(" {} nn score: {} ".format(3,knn.score(x_test,y_test)))
```
