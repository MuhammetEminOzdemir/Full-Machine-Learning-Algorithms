```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
x=iris.data
y=iris.target
```

```python
y
```

```python
x
```

```python
#Değerleri normalize edelim çünkü bir i büyük bir değer küçük olduğundan sapma olmasın 0 1 arasına sığdıralım

#knn algoritmasıında normalize şart

x=(x-np.min(x))/ (np.max(x)-np.min(x))
```

```python
x
```

```python
#train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
#K-fold cross validation

from sklearn.model_selection import cross_val_score

accuracies=cross_val_score(estimator=knn,X=x_train,y=y_train,cv=10)  #estimator demek cross validation yaparken knn algoritmasını kullan demek

#cv=10 cross validation daki k değeri ne olsun
```

```python
accuracies
```

```python
print("Average Accuary:",np.mean(accuracies))
```

```python
print("Average Accuary:",np.std(accuracies)) #ne kadar tutarlı
```

k = 3 değeri gerçekten iyi bir değermiş onu anlaıyoruz

Şimdi ilk başta hiç kullanmaıdğımız test datası ile test edelim

cross calidaton birden fazla accuary değeri ile tutarlılığa bakar

```python
knn.fit(x_train,y_train)
print("Test Accuary:",knn.score(x_test,y_test))
```

```python
#Grid Search Cross Validation 

#knn değeri en iyi 3 mü? onu bulmaya yarar

from sklearn.model_selection import GridSearchCV
```

```python
from IPython.display import display
from PIL import Image

# Fotoğrafı yükleme
img = Image.open("grid.png")

# Fotoğrafı görüntüleme
display(img)
```

```python
grid={"n_neighbors":np.arange(1,50)} #dictionary
knn=KNeighborsClassifier()

knn_cv=GridSearchCV(knn,grid,cv=10)  #gridsearch
knn_cv.fit(x,y)
```

```python
print("Tuned Hyperparameter:",knn_cv.best_params_)  #en iyi parametre seçer
print("Tuned Parametreye göre en iyi accuary :",knn_cv.best_score_)  #en iyi parametre seçer
```

# Grid Search with Logistic Regression

#iris datasındaki ilk 2 class alalım 3 class vardı 

#Çünkü Logistic regression ile binary class olmalı 

```python
x=x[:100,:]
y=y[:100]

#y class larımız onu 2 aldık peki neden x yani featuralarda ilk 100 veriyi aldık çünkü shape boyutu eşit olmalı
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression

grid={"C":np.logspace(-3,3,7),"penalty":["l1","l2"]}  #Buradaki C büyükse overfitting küçükse underfitting olur iyi ayarlanmalı
#l1 =Lasso l2= ridge alg.

logreg=LogisticRegression()
logreg_cv=GridSearchCV(logreg,grid,cv=10)
logreg_cv.fit(x,y) #normalde x_train ve y_train ile fit edilmeli ama amacımız grid serach tahminleme değil
```

```python
print("Tuned Hyperparameter (best parameters):",logreg_cv.best_params_)  #en iyi parametre seçer
print("Tuned Parametreye göre en iyi accuary (best accuary) :",logreg_cv.best_score_)  #en iyi parametre seçer
```
